```python
import pandas as pd
import numpy as np
```

```python
Movie = movie.merge(credit,on ='title')
```

```python
New_movie = Movie[['title','movie_id','genres','keywords','overview','cast','crew']]
```

```python
New_movie.isnull().sum()
```

```python
New_movie.dropna(inplace=True)
```

```python
New_movie['keywords'][0]
```

```python
import ast
```

```python
def convert(text):
    L =[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
        
    return L


```

```python
New_movie['keywords'] = New_movie['keywords'].apply(convert)
```

```python
New_movie.head()
```

```python
New_movie['genres'] = New_movie['genres'].apply(convert)
```

```python
New_movie.head()
```

```python
def converter(text):
    L =[]
    counter = 0
    for i in ast.literal_eval(text):
          if counter <3:
                L.append(i['name'])
                counter+=1
                
    return L
```

```python
New_movie['cast'] = New_movie['cast'].apply(converter)
```

```python
New_movie.head()
```

```python
New_movie['cast'] = New_movie['cast'].apply(lambda x:x[0:3])
```

```python
def convertor(text):
    L =[]
    for i in ast.literal_eval(text):
         if i['job'] == 'Director':
                L.append(i['name'])
                
    return L
```

```python
New_movie['crew'] = New_movie['crew'].apply(convertor)
```

```python
New_movie.head()
```

```python
# Make string into List
New_movie['overview'].apply(lambda x :x.split())
```

```python
New_movie.head()
```

```python
 New_movie['genres'] = New_movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
New_movie['cast'] = New_movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
New_movie['crew'] = New_movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
New_movie['keywords'] = New_movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
New_movie.head()
```

```python
New_movie['Tags'] = New_movie['cast'] + New_movie['crew'] + New_movie['keywords'] + New_movie['genres']
```

```python
New_movie.head()
```

```python
Final = New_movie[['movie_id','title','Tags']]
```

```python
Final.head()
```

```python
 Final['Tags'] = Final['Tags'].apply(lambda x:" ".join(x))
```

```python
Final['Tags'][0]
```

```python
 Final['Tags'] = Final['Tags'].apply(lambda x:x.lower())
```

```python
Final.head()
```

```python
from nltk.stem.porter import PorterStemmer
PS = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(PS.stem(i))
    return " ".join(y)
```

```python
Final['Tags'] =Final['Tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features =4000,stop_words='english')
```

```python
 Vector = CV.fit_transform(Final['Tags']).toarray()
```

```python
Vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
 simi_matrix = cosine_similarity(Vector)
```

```python
# Index position chnage when sorting so apply enumerate
enumerate(simi_matrix[0])
```

```python
# Convert into list
list(enumerate(simi_matrix[0]))
```

```python
# Sorted on basis of second movie so use lambda
sorted(list(enumerate(simi_matrix[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    index = Final[Final['title'] == movie].index[0]
    distances = sorted(list(enumerate(simi_matrix[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(Final.iloc[i[0]].title)
        
    
```
